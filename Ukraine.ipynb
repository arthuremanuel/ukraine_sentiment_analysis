Data Source: https://www.kaggle.com/jpmiller/conflict-in-ukraine?select=conflict_data_ukr.csv

```python
import pandas as pd

news = pd.read_csv('conflict_data_ukr.csv')
news = news.drop(index = 0)
news.head()
```

```python
news['source_article']
```

```python
data = pd.DataFrame()

for n in news['source_article']:
    n = n.replace("\"",'')
    n = n.split(',')
    dic = {
        'source': n[0],
        'date': n[1],
        'headline' : n[2],
    }
    data = data.append(dic,ignore_index=True)
data
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

a = 'This was a good movie.'
sid.polarity_scores(a)
```

```python
a = 'This was the best, most awesome movie EVER MADE!!!'
sid.polarity_scores(a)
```

```python
sent = []
for i in data['headline']:
    sent.append(sid.polarity_scores(i)['compound'])
data['Compound'] = sent
data.head()
```

```python
sent = []
for i in data['Compound']:
    sent.append('POSITIVE' if i >=0 else 'NEGATIVE')
data['Sentiment'] = sent
data.head()
```
